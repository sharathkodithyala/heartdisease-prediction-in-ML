```python
# sepaeratin the labels 
x=dataa.drop(columns='outcome',axis=1)
y=dataa['outcome']
```

```python
print(y)
```

```python
#makeing the standraization
scaler= StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data=scaler.transform(x)
```

```python
print(standardized_data)
```

```python
x=standardized_data
y=dataa['outcome']
```

```python
print(x)
print(y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training th support v
classifier.fit(X_train,Y_train)
```

```python
#model evalution
#accuracy score
X_train_predict=classifier.predict(X_train)
training_dataa= accuracy_score(X_train_predict,Y_train)
```

```python
print(training_dataa)
```

```python
#making the predixting system
inputdata=(0,3,5,3,4,332,43,431)
inputdatanp=np.asarray(inputdata)
inputreshape=inputdatanp.reshape(1,-1)
stddata=scaler.transform(inputreshape)
print(stddata)
prediction=classifier.predict(stddata)
print(prediction)
```

```python
if(prediction[0]==0):
    print("the person is not having diabetes")
else:
    print("the person is diabetic")
```
